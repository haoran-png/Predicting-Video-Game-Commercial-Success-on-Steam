# Initial Basic EDA

***This section serves as a preliminary step before the main exploratory analysis. The goal is to perform light but essential cleaning, specifically targeting missing values and duplicates to lay the groundwork for a deeper, more insightful EDA. That deeper analysis will then close with a comprehensive cleaning phase, leaving the dataset in a tidy, readable state.***

```python
import pandas as pd
from pathlib import Path

script_path = Path.cwd()
data_path = script_path.parent/"data"/"raw"/"games.csv"

df = pd.read_csv(data_path)
df.shape
df.head(5)
```

***From this initial graph, we can immediately spot a data loading issue: the AppID column appears misaligned with its values. Since the raw CSV file shows no such problem, this is a pandas specific artifact, pandas is incorrectly treating one of the columns as the row index, shifting everything out of place.***

***The fix is straightforward, explicitly passing index_col=False when reading the file forces pandas to use its own default integer index, restoring the correct column alignment.***

```python
df = pd.read_csv(data_path, index_col=False)
df.head(5)
```

```python
df.info()

null_counts = df.isnull().sum()
null_pct = ((null_counts / len(df)) * 100).round(2)
null_summary = pd.DataFrame({"null_count": null_counts, "null_pct": null_pct})
null_summary[null_summary["null_count"] > 0].sort_values(by="null_count", ascending=False)
```

***Using `df.info()`, we can see that this dataset contains **122,611 entries across 39 columns**, where each entry represents a game (including duplicates, as the dataset has not yet been cleaned). These 39 columns cover a wide range of metrics describing and evaluating each game, making the dataset considerably larger and more complex than what this project requires.***

***Several columns are unlikely to contribute meaningful value to our analysis. Most notably, `Achievements` contains up to **99.97% null values**, making it nearly unusable in its current state. Other columns also exhibit high rates of nullness, as shown in the null value summary above.***

***Beyond missing data, a number of columns are stored in incorrect data types. For example, `Release date` is stored as `object` rather than `datetime`, and `Average playtime forever` is typed as `object` despite the three other playtime-related columns (`Average playtime two weeks`, `Median playtime forever`, `Median playtime two weeks`) all being stored as `int64`. Additionally, several columns appear to have been assigned the wrong dtype entirely, for instance, `Metacritic score` is stored as `bool` and `Metacritic url` as `int64`, which is clearly inconsistent with the kind of data these fields should hold.***

***All of these issues: duplicate entries, high null columns, and dtype inconsistencies will be addressed in the initial cleaning step.***

```python
print(f"Duplicate rows: {df.duplicated().sum()}")
print(f"Duplicate AppIDs: {df['AppID'].duplicated().sum()}")
```

```python
df.describe()
```

***Before examining the statistics, duplicate checking confirmed that the dataset contains no repeated rows or duplicate `AppID` values, meaning each entry uniquely identifies a game.***

***`df.describe()` then reveals several dtype issues flagged earlier. Columns such as `Metacritic url`, `About the game`, `Developers`, `Notes`, and `Score rank` appear in the numeric summary despite clearly not representing quantitative values, further confirming they were assigned incorrect dtypes during ingestion.***

***Beyond dtype issues, the distributions reveal that the dataset is heavily skewed toward low engagement titles. The median values for `Peak CCU`, `Positive` reviews, `Negative` reviews, `Recommendations`, and all playtime related columns are all **0**, meaning more than half the games in the dataset have essentially no recorded player activity. This is worth keeping in mind during analysis, as a small number of high engagement outliers, such as the game with a Peak CCU of over 1,000,000 will heavily influence any mean based metrics.***

***`Price` ranges from free ($0) to $999.98, with a median of just $2.24, reflecting the prevalence of very cheap or free to play titles on the platform. Similarly, `Required age` has a mean of 0.17 and a maximum of 21, indicating that the vast majority of games carry no age restriction.***
